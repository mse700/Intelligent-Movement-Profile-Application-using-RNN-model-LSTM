### Scaling the Input Data

In order to make the input data robust to outlier, we scale the input using RobustScaler from [Scikitlearn](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.RobustScaler.html)



```python
from sklearn.preprocessing import RobustScaler

scale_columns = ['X', 'Y', 'Z']

scaler = RobustScaler()

scaler = scaler.fit(df_train[scale_columns])

df_train.loc[:, scale_columns] = scaler.transform(df_train[scale_columns].to_numpy())
# df_test.loc[:, scale_columns] = scaler.transform(df_test[scale_columns].to_numpy())
```

### Dataset Creation

LSTM model needs the the dataset where the number of sequences (TimeStep) and the Step size(STEP) needs to be clarified. In this model we have considered 13 inputs as TimeSteps from Step siize of 60. In the other words, we take 13 data in every 60 seconds to train our RNN model.  

```python
from scipy import stats

def create_dataset(X, y, time_steps=1, step=1):
    Xs, ys = [], []
    for i in range(0, len(X) - time_steps, step):
        v = X.iloc[i:(i + time_steps)].values
        labels = y.iloc[i: i + time_steps]
        Xs.append(v)        
        ys.append(stats.mode(labels)[0][0])
    return np.array(Xs), np.array(ys).reshape(-1, 1)

TIME_STEPS = 13
STEP = 60

X_train, y_train = create_dataset(
    df_train[['X', 'Y', 'Z']], 
    df_train.Activity_Label, 
    TIME_STEPS, 
    STEP
)
```

```python
# Using OneHotEncoding for transforming string values to numerical ones
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore', sparse=False)

enc = enc.fit(y_train)

y_train = enc.transform(y_train)
# y_test = enc.transform(y_test)
```

### Model with LSTM layer 

```python
model = Sequential()
model.add(
    keras.layers.Bidirectional(
      keras.layers.LSTM(
          units=128, 
          input_shape=[X_train.shape[1], X_train.shape[2]]
      )
    )
)
model.add(keras.layers.Dropout(rate=0.5))
model.add(keras.layers.Dense(units=128, activation='relu'))
model.add(keras.layers.Dense(y_train.shape[1], activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
```

```python
history = model.fit(X_train, y_train,epochs=20,batch_size=64,validation_split=0.1,shuffle=True)
```

```python
from sklearn.metrics import confusion_matrix

def plot_cm(y_true, y_pred, class_names):
  cm = confusion_matrix(y_true, y_pred)
  fig, ax = plt.subplots(figsize=(18, 16)) 
  ax = sns.heatmap(
      cm, 
      annot=True, 
      fmt="d", 
      cmap=sns.diverging_palette(220, 20, n=7),
      ax=ax
  )

  plt.ylabel('Actual')
  plt.xlabel('Predicted')
  ax.set_xticklabels(class_names)
  ax.set_yticklabels(class_names)
  b, t = plt.ylim() # discover the values for bottom and top
  b += 0.5 # Add 0.5 to the bottom
  t -= 0.5 # Subtract 0.5 from the top
  plt.ylim(b, t) # update the ylim(bottom, top) values
  plt.show() # ta-da!
```
